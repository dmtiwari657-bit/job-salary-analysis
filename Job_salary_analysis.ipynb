```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/job salary prediction.zip")
df.head()
```

```python
df.shape
df.columns
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.describe()
```

```python
# Observation
# “The salary values vary significantly across the dataset.”
# “Experience and skills show variation among individuals.”
```

```python
# Category Analysis
df['job_title'].value_counts()
```

```python
df['education_level'].value_counts()
```

```python
# Observation:
# “Some job roles are more frequent than others.”
# “Most candidates belong to a specific education level.”
```

```python
# Experience vs Salary
df.groupby('experience_years')['salary'].mean()
```

```python
# Observation
# “Salary increases as experience increases.”
# “More experienced individuals tend to earn higher salaries.”
```

```python
# Education vs Salary
df.groupby('education_level')['salary'].mean()
```

```python
# Observation
# “Higher education levels are associated with higher salaries.”
# “Education plays an important role in salary determination.”
```

```python
# Skill Count vs Salary
df.groupby('skills_count')['salary'].mean()
```

```python
# Observation
# “Salary tends to increase with the number of skills.”
# “Individuals with more skills earn higher salaries on average.”
```

```python
# Company Size vs Salary
df.groupby('company_size')['salary'].mean()
```

```python
# Observation
# “Employees in larger companies tend to earn higher salaries.”
# “Company size has an impact on salary levels.”
```

```python
# Remote Work vs Salary
df.groupby('remote_work')['salary'].mean()
```

```python
# Observation
# “There is a difference in salary between remote and non-remote jobs.”
# “Work mode may influence salary structure.”
```

```python
# Job Role vs Salary
df.groupby('job_title')['salary'].mean()
```

```python
# Observation
# “Different job roles have significantly different salary levels.”
# “Technical roles tend to have higher salaries.”
```

```python
# Insights
# “Salary varies significantly across individuals in the dataset.”
# “Higher experience is associated with higher salary.”
# “Education level has a positive impact on salary.”
# “More skills generally lead to higher salary.”
# “Salary differs across job roles.”
# “Company size and work type (remote/on-site) also influence salary.”
```

```python
# Visualization
# Salary Distribution (Histogram)
import matplotlib.pyplot as plt

df['salary'].plot(kind='hist')
plt.title("Salary Distribution")
plt.show()
```

```python
# Observation
# “Most salaries are concentrated in a specific range.”
# “The distribution is slightly skewed.”
```

```python
#Experience vs Salary(Trend)
plt.scatter(df['experience_years'], df['salary'])
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Observation
# “Salary tends to increase with experience.”
# “There is a positive relationship between experience and salary.”
```

```python
# Job Role Count(Bar Graph)
df['job_title'].value_counts().plot(kind='bar')
plt.title("Job Role Distribution")
plt.show()
```

```python
# Observation
# “Few job roles dominate the dataset.”
```

```python
# Conclusion
# “The dataset was cleaned and analyzed successfully.”
# “Experience and skills play a major role in salary prediction.”
# “Visualizations helped in identifying trends and patterns.”
# “The model shows potential but can be improved further.”
```
